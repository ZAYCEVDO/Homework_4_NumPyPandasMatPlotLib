```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random 
import timeit
import math
%matplotlib inline
```

```python
#Наивное скалярное произведение векторов
def scalarMultNaive (x, y):
    assert x.ndim == 1 and y.ndim == 1
    assert x.size == y.size
    result = 0
    for i in range(x.size):
        result += x[i]*y[i]
    return result

#Скалярное произведение NumPy
def scalarMultNp (x, y):
    assert x.ndim == 1 and y.ndim == 1
    assert x.size == y.size
    return np.dot(x, y)
```

```python
#Cоздание векторов
x10 = np.array(random.sample(range (100), 10))
y10 = np.array(random.sample(range (100), 10))

x50 = np.array(random.sample(range (100), 50))
y50 = np.array(random.sample(range (100), 50))

x100 = np.array(random.sample(range (100), 100))
y100 = np.array(random.sample(range (100), 100))   
```

```python
#Вычисление наивных скалярных произведений
start_time = timeit.default_timer()
scalarMultNaive (x10, y10)
time10 = timeit.default_timer() - start_time

start_time = timeit.default_timer()
scalarMultNaive (x50, y50)
time50 = timeit.default_timer() - start_time

start_time = timeit.default_timer()
scalarMultNaive (x100, y100)
time100 = timeit.default_timer() - start_time
```

```python
#Вычисление скалярных произведений NumPy
start_time = timeit.default_timer()
scalarMultNp (x10, y10)
nptime10 = timeit.default_timer() - start_time

start_time = timeit.default_timer()
scalarMultNp (x50, y50)
nptime50 = timeit.default_timer() - start_time

start_time = timeit.default_timer()
scalarMultNp (x100, y100)
nptime100 = timeit.default_timer() - start_time
```

```python
timelist = [time10, time50, time100]
nptimelist = [nptime10, nptime50, nptime100]
vectorlengthlist = [10, 50, 100]
```

```python
#Построение графиков
plt.style.use('classic')
plt.title('Cравнение времени выполнения скалярных произвeдений')
plt.xlabel('Длина вектора')
plt.ylabel('Время выполнения')
plt.grid()
plt.plot(vectorlengthlist, timelist, '-', color = 'g', label = "Наивное")
plt.plot(vectorlengthlist, nptimelist, '-', color = 'r', label = "NumPy")
plt.legend()
```

```python
#Сохранение результатов в DataFrame
results = pd.DataFrame({
    "Длина вектора": vectorlengthlist,
    "Наивное время(sec)": timelist,
    "Время NumPy(sec)": nptimelist
})
print(results)
```
